# Full Evaluation And Comparsion
Evaluate all trained models aginst all error functions

```python

data = {'Model':[], 'Training Loss Function':[], 'Epochs':[], 'Hidden Layer Size':[], 'Dropout Ratio':[], 
        'Features Number':[], 'MASE':[], 'MAE': [], 'SMAPE': [], 'RMSE': [] }

models_arch = ['1-LSTM', '2-LSTM', '3-LSTM']
loss_functions = [mase, smape, rmse, mae]

MODEL_BASE_DIR = '../models'
TRAINING_DATA_PATH = '../Dataset/Train/Hourly-train.csv'
TEST_DATA_PATH = '../Dataset/Test/Hourly-test.csv'

for model_arc in models_arch:
    s = f'{MODEL_BASE_DIR}/{model_arc}/*/'
    models = glob(s)
    for model_path in models:
        
        model_results = {'training_error': [], 'test_error': [], 'validation_error': [] }
        
        for loss in loss_functions:
            hyperparameters, training_error, test_error, validation_error = load_and_evaluate_model(model_path, 
                                                                                                    TRAINING_DATA_PATH, 
                                                                                                    TEST_DATA_PATH, loss)
            model_results['training_error'].append(training_error)
            model_results['test_error'].append(test_error)
            model_results['validation_error'].append(validation_error)
        
        data['Model'].append(model_arc)
        data['Training Loss Function'].append(hyperparameters['loss'])
        data['Epochs'].append(hyperparameters['epochs'])
        data['Hidden Layer Size'].append(hyperparameters['hidden_layer_size'])
        data['Dropout Ratio'].append(hyperparameters['dropout_ratio'])
        data['Features Number'].append(hyperparameters['features_number'])
        data['MASE'].append(model_results['training_error'][0])
        data['SMAPE'].append(model_results['training_error'][1])
        data['RMSE'].append(model_results['training_error'][2])
        data['MAE'].append(model_results['training_error'][3])

df = pd.DataFrame(data=data)

dfStyler = df.style.set_properties(**{'text-align': 'center'})
dfStyler.set_table_styles([dict(selector='th', props=[('text-align', 'center')])])
```
